```python
# While Loop always required a statement along with it in order to run

count = 0
while count<10:
   print('The count is:', count)
   count = count + 1

print("Ended While Loop")
```

```python
# Example of infinite for loop
count=100
while True:
    print("The count is :",count)
```

```python
# Breaking While Loop Using If Condition
count=0
while True:
    print("count : ",count)
    if count ==10:
        break
    else:
        count+=1
        
```
